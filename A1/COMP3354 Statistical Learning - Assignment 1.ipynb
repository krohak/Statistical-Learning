## (a)

```r
mpg_median = median(mpg)
mpg_median
```

```r
obs = dim(Auto)[1]
```

```r
mpg01 = rep(0, obs)
mpg01[mpg > mpg_median] = 1
```

```r
df = data.frame(Auto,mpg01)
```

```r
tail(df)
```

## (b) 

```r
options(repr.plot.width=10, repr.plot.height=10)
plot(df)
options(repr.plot.width=5, repr.plot.height=5)
```

```r
par(mfrow=c(2,3))
boxplot(displacement ~ mpg01, col=c("red", "blue"))
boxplot(horsepower ~ mpg01, col=c("red", "blue"))
boxplot(weight ~ mpg01, col=c("red", "blue"))
boxplot(acceleration ~ mpg01, col=c("red", "blue"))
boxplot(year ~ mpg01, col=c("red", "blue"))
```

#### From the scatterplots and boxplots above, all features show some sort of relationship with mpg01. The most useful features to predict mpg01 seem to be weight and horsepower.

## (c)

```r
rows = sample(x=nrow(df), size=.80*nrow(df))
train =  df[rows, ]
test = df[-rows, ]
```

```r
head(train)
```

```r
head(test)
```

#### The data is split into 80-20 training-testing datasets.

## (d)

#### Using only Horsepower and Weight as the predictors:

```r
glm.fit = glm(mpg01 ~ horsepower+weight, family=binomial, data=train)
```

```r
summary(glm.fit)
```

```r
glm.probs = predict(glm.fit, test, type="response")
```

```r
glm.pred  = rep(0, dim(test)[1])
glm.pred[glm.probs >.5] = 1 
```

```r
table(glm.pred, test$mpg01)
```

#### Test Accuracy 

```r
mean(glm.pred==test$mpg01)
```

#### Test Error 

```r
mean(glm.pred!=test$mpg01)
```

#### The test error is 13% with only Horsepower and Weight as the predictors.

#### Logistic Regresssion with all the features as predictors:

```r
glm.fit = glm(mpg01 ~ displacement+horsepower+weight+acceleration+year+cylinders+origin, family=binomial, data=train)
```

```r
glm.probs = predict(glm.fit, test, type="response")
glm.pred  = rep(0, dim(test)[1])
glm.pred[glm.probs >.5] = 1 
```

#### Test Error 

```r
mean(glm.pred!=test$mpg01)
```

#### The test error decreases from 13% to 5% when training the model with all the features.

## (b)

```r
std_err_hat = sd(medv) / sqrt(dim(Boston)[1])
std_err_hat
```

#### standard error of μ^ = 0.408861147497535 

## (c)

```r
boot.fn = function(data, index) {
    mu = mean(data[index])
    return (mu)
}

boot(medv, boot.fn, 1000)
```

#### The std err of μ^ computed using Bootstrap is 0.4144862 which is close to our estimate above, 0.408861147497535 

## (d)

Using the formula [μˆ − 2SE(μˆ), μˆ + 2SE(μˆ)], the confidence intervals are:

```r
c( u_hat - 2*std_err_hat, u_hat + 2*std_err_hat )
```

Using t.test():

```r
t.test(medv)
```

#### The 95 percent confidence interval using the formula are [21.7150840291156, 23.3505286191057] and using t.test() are [21.72953, 23.33608]. Both methods yield similar C.I.
