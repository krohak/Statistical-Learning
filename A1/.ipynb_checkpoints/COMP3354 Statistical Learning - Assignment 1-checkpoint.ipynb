```r
glm.fit = glm(mpg01~horsepower+weight, family=binomial, subset=train)
```

```r
summary(glm.fit)$coef
```

```r
glm.probs = predict(glm.fit, Auto.test, type="response")
```

```r
test_len = obs - sum(train)
```

```r
glm.pred  =rep(0, test_len)
glm.pred[glm.probs >.5] = 1 
```

```r
table(glm.pred , mpg01.test)
```

** Test Accuracy **

```r
mean(glm.pred==mpg01.test)
```

** Test Error **

```r
mean(glm.pred!=mpg01.test)
```

** The test error is 20% **

```r
set.seed(2475)
library(ISLR)
attach(Default)
head(Default)
```

## (a)

```r
fit.glm = glm(default ~ income + balance, family = "binomial")
summary(fit.glm)
```

** Thus, the estimated standard errors for the coefficients associated with income and balance are 4.985e-06 and 2.274e-04 respectively **

## (b)

```r
boot.fn = function(data, index) {
    fit = glm(default ~ income + balance, data = data, family = "binomial", subset = index)
    return (coef(fit))
}
```

## (c)

```r
library(boot)
boot(Default, boot.fn, 1000)
```

** The standard errors of the logistic regression coefficients for income and balance are 4.842083e-06 and 2.319733e-04 respectively. **

## (d) 

** Both methods yield estimated standard errors that are pretty close. **

```r
set.seed(2475)
library(MASS)
attach(Boston)
```

```r
head(Boston)
```

## (a)

```r
u_hat = mean(medv)
u_hat
```

** μˆ = 22.5328063241107 **

## (b)

```r
std_err_hat = sd(medv) / sqrt(dim(Boston)[1])
std_err_hat
```

** standard error of μ^ = 0.408861147497535 ** 

## (c)

```r
boot.fn = function(data, index) {
    mu = mean(data[index])
    return (mu)
}

boot(medv, boot.fn, 1000)
```

** The std err of μ^ computed using Bootstrap is 0.4144862 which is close to our estimate above, 0.408861147497535 **

## (d)

Using the formula [μˆ − 2SE(μˆ), μˆ + 2SE(μˆ)], the confidence intervals are:

```r
c( u_hat - 2*std_err_hat, u_hat + 2*std_err_hat )
```

Using t.test():

```r
t.test(medv)
```

** The 95 percent confidence interval using the formula are [21.7150840291156, 23.3505286191057] and using t.test() are [21.72953, 23.33608]. Both methods yield similar C.I. **
